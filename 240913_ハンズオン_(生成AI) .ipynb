```python
#今回はAPIキーを直接入力するために、このようなおまじないをいれます。これを走らせると、キーの入力が求めれます。
import os
import openai
from getpass import getpass

# APIキーを安全に入力
api_key = getpass("OpenAIのAPIキーを入力してください: ")
os.environ["OPENAI_API_KEY"] = api_key
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
from openai import OpenAI
# chatGPTに質問する
# OpenAIクライアントの初期化
client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini", # あなたのデプロイメント名に変更してください
    messages=[
        {"role": "system", "content": "あなたは自動車エンジニアのアシスタントです。"},
        {"role": "user", "content": "Hondaの最新の安全技術について教えてください。"}
    ]
)
# "content"を違うものに変えてみる

# 回答の表示
print(response.choices[0].message.content)
```

```python
from openai import OpenAI

# OpenAIクライアントの初期化
client = OpenAI()

# 質問の設定
question = "本田技研工業の創業者は誰ですか？"

try:
    # ChatGPTへのリクエスト
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # または利用可能な最新のモデル
        messages=[
            {"role": "system", "content": "あなたは優秀なAIアシスタントです。"},
            {"role": "user", "content": question}
        ]
    )
    # "content"の"あなたは優秀なAIアシスタントです。"を"あなたはホンダのエンジニアです。実は創業者は複数いることを知っています"に変えてみる

    # 回答の取得
    answer = response.choices[0].message.content

    # 結果の表示
    print(f"質問: {question}")
    print(f"回答: {answer}")

except Exception as e:
    print(f"エラーが発生しました: {str(e)}")
```

### ③-1プロンプトエンジニアリングの応用
プロンプトエンジニアリングとは、AIモデルから望ましい回答を得るために、適切な指示（プロンプト）を設計することです。以下に基本的なテクニックを紹介します。
- 明確な指示: AIに具体的に何をしてほしいか明確に伝える
- コンテキストの提供: 必要な背景情報を与える
- 出力フォーマットの指定: 望ましい回答の形式を指定する
- ステップバイステップの指示: 複雑なタスクを小さなステップに分ける

```python
#まずはGPT-4o-miniのおさらい
def ask_gpt4o(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # GPT-4oのモデル名
        messages=[
            {"role": "system", "content": "あなたは優秀なAIアシスタントです。"},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content

question = "本田技研工業の創業者は誰ですか？"
# question = "ここの部分"をいじってみる

answer = ask_gpt4o(question)
print(f"質問: {question}")
print(f"回答: {answer}")
```

## ③-2プロンプトエンジニアリングの応用例（スキップしてもOK）
ここでは、プログラミングに近いような記載をしておくことで、より柔軟に、より適切な回答を得ることができることを知ります。
知っておくだけで十分です！

```python
def engineer_prompt(task, context, format_instruction, steps):
    prompt = f"""
タスク: {task}
コンテキスト: {context}
出力フォーマット: {format_instruction}
ステップ:
{steps}

上記の指示に従って回答してください。
"""
    return ask_gpt4o(prompt)

task = "本田技研工業の新製品のプレスリリースを作成する"
context = "本田技研工業は、新しい電気自動車「Honda e:NS2」を発表しました。この車は最新のAI技術を搭載し、自動運転機能を備えています。"
format_instruction = "プレスリリースは、タイトル、概要、特徴、価格、発売日の順で構成してください。"
steps = """
1. プレスリリースのタイトルを考える
2. 製品の概要を2-3文で説明する
3. 主要な特徴を3つ箇条書きで列挙する
4. 価格情報を記載する
5. 発売予定日を記載する
"""

# task = "ここの部分"をいじってみる
# context = "ここの部分"をいじってみる
# format_instruction = "ここの部分"をいじってみる
# steps = "ここの部分"をいじってみる

press_release = engineer_prompt(task, context, format_instruction, steps)
print(press_release)
```

## ④-1プロンプトの最適化
プロンプトを最適化するには、試行錯誤が必要です。以下の例で、プロンプトを少しずつ改善していく過程を見てみましょう。

```python
def optimize_prompt(base_prompt, improvements):
    results = []
    current_prompt = base_prompt

    for i, improvement in enumerate(improvements, 1):
        current_prompt += f"\n{improvement}"
        response = ask_gpt4o(current_prompt)
        results.append(f"改善{i}:\n{response}\n")

    return "\n".join(results)

base_prompt = "本田技研工業の歴史について教えてください。"
improvements = [
    "特に創業時の話に焦点を当ててください。",
    "創業者の本田宗一郎氏の性格や信念についても触れてください。",
    "会社の成長に影響を与えた主要な出来事を2つ挙げてください。"
]
# base_prompt = "ここの部分"をいじってみる
# improvements = "ここの部分"をいじってみる

optimization_results = optimize_prompt(base_prompt, improvements)
print(optimization_results)
```
